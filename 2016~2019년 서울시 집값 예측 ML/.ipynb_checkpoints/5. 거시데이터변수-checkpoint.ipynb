```python
import pandas as pd
import numpy as np
```

```python
CPIdf = pd.read_csv('data/CPI_monthly.csv',skiprows=1, encoding='cp949',usecols=['시점','총지수'])
CPIdf.rename(columns={'총지수':'소비자물가지수'}, inplace = True)
CPIdf['연도'] = CPIdf.copy()["시점"].str.split(".", expand=True)[0]
CPIdf['월'] = CPIdf.copy()["시점"].str.split(".", expand=True)[1]
CPIdf.drop(['시점'], axis=1, inplace=True)

CPIdf
```

```python
GDPdf = pd.read_csv('data/GDP_qtr.csv', encoding='cp949')
GDPdf.rename(columns={'국내총생산(시장가격 GDP)':'GDP','분기':'월'}, inplace = True)
GDPdf['연도'] = GDPdf.copy()["시점"].str.split(" ", expand=True)[0]
GDPdf['월'] = GDPdf.copy()["시점"].str.split(" ", expand=True)[1]
GDPdf.drop(['시점'], axis=1, inplace=True)

GDPdf
```

```python

KOSPIdf = pd.read_csv('data/Kospi_monthly.csv', encoding='cp949',usecols=['시점','평균'])
KOSPIdf.rename(columns={'평균':'평균KOSPI지수'}, inplace = True)
KOSPIdf['연도'] = KOSPIdf.copy()["시점"].str.split(".", expand=True)[0]
KOSPIdf['월'] = KOSPIdf.copy()["시점"].str.split(".", expand=True)[1]
KOSPIdf.drop(['시점'], axis=1, inplace=True)
KOSPIdf = KOSPIdf.dropna()
KOSPIdf
```

```python
LPdf = pd.read_csv('data/LP_monthly.csv', encoding='cp949')
LPdf.rename(columns={'L (광의 유동성)':'총유동성'}, inplace = True)
LPdf['연도'] = LPdf.copy()["시점"].str.split(".", expand=True)[0]
LPdf['월'] = LPdf.copy()["시점"].str.split(".", expand=True)[1]
LPdf.drop(['시점'], axis=1, inplace=True)
LPdf = LPdf.dropna()
LPdf
```

```python
MorgageIdf = pd.read_csv('data/morgage_monthly.csv', encoding='cp949',usecols=('시점','주택담보대출'))
MorgageIdf['연도'] = MorgageIdf.copy()["시점"].str.split(".", expand=True)[0]
MorgageIdf['월'] = MorgageIdf.copy()["시점"].str.split(".", expand=True)[1]
MorgageIdf.drop(['시점'], axis=1, inplace=True)
MorgageIdf = MorgageIdf.dropna()
MorgageIdf
```

```python
USDEXdf = pd.read_csv('data/USD_EX_monthly.csv', encoding='cp949')
USDEXdf.rename(columns={'아시아':'환율'}, inplace = True)
USDEXdf['연도'] = USDEXdf.copy()["시점"].str.split(".", expand=True)[0]
USDEXdf['월'] = USDEXdf.copy()["시점"].str.split(".", expand=True)[1]
USDEXdf.drop(['시점'], axis=1, inplace=True)
USDEXdf = USDEXdf.dropna()
USDEXdf['환율'] = USDEXdf.copy()["환율"].str.split(".", expand=True)[0]
USDEXdf["환율"] = USDEXdf.copy()["환율"].str.replace(r'\D', '')

USDEXdf
```
