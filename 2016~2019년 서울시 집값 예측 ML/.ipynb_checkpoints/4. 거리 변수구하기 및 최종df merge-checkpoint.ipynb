```python
#유치원, 어린이집def

def search_PS3(lon, lat):
    dfd = pd.DataFrame(columns = ['distance'])
    page = 1
    while True:
        
        url = "https://dapi.kakao.com/v2/local/search/category.json?&category_group_code=PS3&x="\
        +str(lon)+"&y="+str(lat)+"&page="+str(page)+"&radius=1500"
        json_obj = requests.get(urlparse(url).geturl(),headers={"Authorization":"KakaoAK e0a854d712dcc8a939b48a11d0201978"}).json()
        for document in json_obj['documents']:
            dfd_s = pd.DataFrame(document, index=[0])[['distance']]
            dfd = dfd.append(dfd_s)
        if json_obj['meta']['is_end'] == False:
            page += 1
        else:
            return dfd
```

```python
list = []

for row in propdf.itertuples():
    lon = getattr(row, 'lon')
    lat = getattr(row, 'lat')
    list.append(search_PS3(lon, lat))
    
df_PS3 = pd.DataFrame(list, columns =[ 'distance'])
df_PS3.head()

```

```python
df_PS3.to_csv("df_PS3.csv",mode='w')
```

```python
#csv로 저장한 각 카테고리별 거리 불러오기

dfps3 = pd.read_csv("df_PS3.csv", encoding='utf-8',index_col = 0)
dfsc4 = pd.read_csv("df_SC4.csv", encoding='utf-8',index_col = 0)
dfct1 = pd.read_csv("df_CT1.csv", encoding='utf-8',index_col = 0)
dfhp8 = pd.read_csv("df_HP8.csv", encoding='utf-8',index_col = 0)
dfmt1 = pd.read_csv("df_MT1.csv", encoding='utf-8',index_col = 0)
dfp03 = pd.read_csv("df_PO3.csv", encoding='utf-8',index_col = 0)
```

```python
#dataframe을 list로 변환후 List Split하기


listps = dfps3.distance.str.split('\n0   ')
listsc = dfsc4.distance.str.split('\n0  ')
listct = dfct1.distance.str.split('\n0   ')
listhp = dfhp8.distance.str.split('\n0   ')
listmt = dfmt1.distance.str.split('\n0  ')
listp = dfp03.distance.str.split('\n0   ')
listp
```

```python
# list 순서대로 한뒤 첫번째 column drop

dfps = pd.DataFrame.from_records(listps)
dfps = dfps.copy().drop(0,axis=1)
dfsc = pd.DataFrame.from_records(listsc)
dfsc = dfsc.copy().drop(0,axis=1)
dfct = pd.DataFrame.from_records(listct)
dfct = dfct.copy().drop(0,axis=1)
dfhp = pd.DataFrame.from_records(listhp)
dfhp = dfhp.copy().drop(0,axis=1)
dfmt = pd.DataFrame.from_records(listmt)
dfmt = dfmt.copy().drop(0,axis=1)
dfp = pd.DataFrame.from_records(listp)
dfp = dfp.copy().drop(0,axis=1)
```

```python
# str 리스트를 int로 변환


dfps = dfps.apply(pd.to_numeric)
dfsc = dfsc.apply(pd.to_numeric)
dfct = dfct.apply(pd.to_numeric)
dfhp = dfhp.apply(pd.to_numeric)
dfmt = dfmt.apply(pd.to_numeric)
dfp = dfp.apply(pd.to_numeric)
```

```python
dfps
```

```python
#거리내 개수 구하기

dfpscnt = dfps[dfps<1500].count(axis=1)
dfsccnt = dfsc[dfsc<2000].count(axis=1)
dfctcnt = dfct[dfct<3000].count(axis=1)
dfhpcnt = dfhp[dfhp<3000].count(axis=1)
dfmtcnt = dfmt[dfmt<1500].count(axis=1)
dfpcnt = dfp[dfp<1500].count(axis=1)
dfsccnt
```

```python
#Count한 거리를 df로 저장

dfps1 = pd.DataFrame(dfpscnt)
dfsc1 = pd.DataFrame(dfsccnt)
dfct1 = pd.DataFrame(dfctcnt)
dfhp1 = pd.DataFrame(dfhpcnt)
dfmt1 = pd.DataFrame(dfmtcnt)
dfp1 = pd.DataFrame(dfpcnt)
dfp1
```

```python
#count한 거리 DF를 column name 변경 및 개수와 최단거리 merge

dfpsdf = pd.merge(dfps1,dfps,left_index=True,right_index=True,how='left')
dfpsdf = dfpsdf.iloc[:,0:2]
dfpsdf = dfpsdf.rename(columns={0:'1.5km 이내 유치원, 어린이집 개수',1:'유치원, 어린이집 최단거리'})

dfscdf = pd.merge(dfsc1,dfsc,left_index=True,right_index=True,how='left')
dfscdf = dfscdf.iloc[:,0:2]
dfscdf = dfscdf.rename(columns={0:'2km 이내 학교 개수',1:'학교 최단거리'})

dfctdf = pd.merge(dfct1,dfct,left_index=True,right_index=True,how='left')
dfctdf = dfctdf.iloc[:,0:2]
dfctdf = dfctdf.rename(columns={0:'2km이내 문화시설 개수',1:'문화시설 최단거리'})

dfhpdf = pd.merge(dfhp1,dfhp,left_index=True,right_index=True,how='left')
dfhpdf = dfhpdf.iloc[:,0:2]
dfhpdf = dfhpdf.rename(columns={0:'2km이내 병원 개수',1:'병원 최단거리'})

dfmtdf = pd.merge(dfmt1,dfmt,left_index=True,right_index=True,how='left')
dfmtdf = dfmtdf.iloc[:,0:2]
dfmtdf = dfmtdf.rename(columns={0:'1.5km이내 공공기관 개수',1:'공공기관 최단거리'})

dfpdf = pd.merge(dfp1,dfp,left_index=True,right_index=True,how='left')
dfpdf = dfpdf.iloc[:,0:2]
dfpdf = dfpdf.rename(columns={0:'1.5km이내 대형마트 개수',1:'대형마트 최단거리'})

dfpsdfdfscdfdfctdfdfhpdfdfmtdfdfpdf

dfpdf
```

```python
#이전에 만들어 놓은 지하철 최단 거리 및 개수 불러오기

dfswdf = pd.read_csv('df_SW8.csv',index_col = 0)
dfswdf
```

```python
#거리 df concat

dfs = pd.concat([dfpsdf,dfscdf,dfctdf,dfhpdf,dfmtdf,dfpdf,dfswdf],axis =1)
dfs
```

```python
# 주소 df 불러오기

dfd = pd.read_csv('df_duplicate.csv', encoding='utf-8', usecols = ['주소'])
dfd
```

```python
# 동일한 index순을 갖고 있기때문에 concat을 통해 병합

dfd1 = pd.concat([dfd,dfs],axis=1)
dfd1
```
