```python
def address_to_latlon(address):
    url = "https://dapi.kakao.com/v2/local/search/address.json?&query=" + address
    result = requests.get(urlparse(url).geturl(),
                          headers={"Authorization":"KakaoAK e0a854d712dcc8a939b48a11d0201978"})
    json_obj = result.json()
    for document in json_obj['documents']:
        global val
        val = [document['address_name'], document['x'], document['y']]
        
    return val
```

```python
list = []
for address in df_new['주소']:
    list.append(address_to_latlon(address))
    
df_new1 = pd.DataFrame(list, columns =[ 'address_name', 'lon', 'lat'])
df_new1.head()

```

```python
df_new1.to_csv('addresstolonlat.csv', mode = 'w')
```

```python
df_new2 = pd.concat([df_new,df_new1],axis=1)
df_new2
```

```python
def search_SW8(lon, lat):
    dfd = pd.DataFrame(columns = ['distance'])
    page = 1
    while True:
        
        url = "https://dapi.kakao.com/v2/local/search/category.json?&category_group_code=SW8&x="\
        +str(lon)+"&y="+str(lat)+"&page="+str(page)+"&radius=5000"
        json_obj = requests.get(urlparse(url).geturl(),headers={"Authorization":"KakaoAK e0a854d712dcc8a939b48a11d0201978"}).json()
        for document in json_obj['documents']:
            dfd_s = pd.DataFrame(document, index=[0])[['distance']]
            dfd = dfd.append(dfd_s)
        if json_obj['meta']['is_end'] == False:
            page += 1
        else:
            return dfd
```

```python
list = []

for row in df_new2.itertuples():
    lon = getattr(row, 'lon')
    lat = getattr(row, 'lat')
    list.append(search_SW8(lon, lat))
    
df_new3 = pd.DataFrame(list, columns =[ 'distance'])
df_new3.head()

```
