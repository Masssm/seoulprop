# 기존 거래내역 데이터에 동데이터 전처리 후 변수 추가 


```python

import pandas as pd
import numpy as np
import glob
prop = pd.read_csv("prop_concat2015~2020.csv", encoding='utf-8', index_col = 0)
prop.head()
```

```python
#서울시 주민등록인구 (동별) 통계
#https://data.seoul.go.kr/dataList/10043/S/2/datasetView.do

df1 = pd.read_csv('seoul_data/seoul_dt1.txt',delimiter = '\t', skipinitialspace=True, usecols = 
                  ['기간','자치구','행정동','인구.3','인구.6','세대당인구'])
df1.rename(columns={'행정동':'동', '인구.3' : '내국인 인구', '인구.6': '외국인 인구'}, inplace = True)

df1
```

```python
#서울시 고령자현황 (동별) 통계
#https://data.seoul.go.kr/dataList/10020/S/2/datasetView.do


df2 = pd.read_csv('seoul_data/seoul_dt2.txt',delimiter = '\t', skipinitialspace=True, usecols = ['기간','자치구','동','65세이상 인구','65세이상 인구.6'])
df2.rename(columns={'65세이상 인구':'65세이상 내국인 인구', '65세이상 인구.6' : '65세이상 외국인 인구'}, inplace = True)
df2
```

```python
#서울시 국민기초생활보장 수급자 (동별) 통계
#https://data.seoul.go.kr/dataList/10113/S/2/datasetView.do


df4 = pd.read_csv('seoul_data/seoul_dt4.txt',delimiter = '\t', skipinitialspace=True, 
                  usecols = ['기간','자치구','동','총 수급자'])
df4
```

```python
#서울시 인구밀도 (동별) 통계
#https://data.seoul.go.kr/dataList/10584/S/2/datasetView.do


df5 = pd.read_csv('seoul_data/seoul_dt5.txt',delimiter = '\t', skipinitialspace=True, 
                  usecols = ['기간','자치구','동','면적','인구밀도'])
df5['기간'] = df5.copy()['기간'].astype(str)
df5
```

```python
#서울시 독거노인 현황 (성별/동별) 통계
#https://data.seoul.go.kr/dataList/10178/S/2/datasetView.do


df6 = pd.read_csv('seoul_data/seoul_dt6.txt',delimiter = '\t', skipinitialspace=True, usecols = 
                  ['기간','자치구','동','합계'])
df6.rename(columns={'합계' : '독거노인 합계'}, inplace = True)
df6
```

```python
#서울시 출생사망(동별) 통계
#https://data.seoul.go.kr/dataList/425/S/2/datasetView.do


df7 = pd.read_csv('seoul_data/seoul_dt7.txt',delimiter = '\t', skipinitialspace=True, usecols = 
                  ['기간','자치구','행정동','출생','사망'])
df7.rename(columns={'행정동':'동'}, inplace = True)
df7
```

```python
#서울시 의료기관 (동별) 통계
#https://data.seoul.go.kr/dataList/10123/S/2/datasetView.do


df8 = pd.read_csv('seoul_data/seoul_dt8.txt',delimiter = '\t', skipinitialspace=True, usecols = 
                  ['기간','자치구','동','합계'])
df8.rename(columns={'합계':'의료기관 합계'}, inplace = True)
df8['기간'] = df8.copy()['기간'].astype(str)

df8
```

```python
#서울시 동 공무원 (정원/동별) 통계
#https://data.seoul.go.kr/dataList/10159/S/2/datasetView.do


df10 = pd.read_csv('seoul_data/seoul_dt10.txt',delimiter = '\t')
df10.rename(columns={'계':'동 공무원 합계'}, inplace = True)
df10['기간'] = df10.copy()['기간'].astype(str)
df10.info()
```

```python
#전처리한 dfs를 merge하고 다시 전처리

dfs = df1.merge(df2,on=('기간','동','자치구'),how='left').merge(df4,on=('기간','동','자치구'),how='left').fillna('void').merge(df5,on=('기간','동','자치구'),how='left').merge(df6,on=('기간','동','자치구'),how='left').merge(df7,on=('기간','동','자치구'),how='left')
dfs
```

```python
#dfs null value check

dfs.isnull().sum()
```

```python
#null값 drop

dfs['면적'].dropna(axis=0)
dfs['인구밀도'].dropna(axis=0)
dfs['독거노인 합계'].dropna(axis=0)
```

```python
#마포2동 마포3동을 숫자를 제거한뒤 마포동으로 변환을 함. 
#기존의 부동산 거래 데이터에는 마포2동 마포3동 모두 마포동으로 되어있기 때문
#동 column에 존재하는 불필요한 row 삭제

dfs['동'] = dfs.copy()['동'].str.replace('\d+', '')
dfs = dfs[((dfs['자치구']!='합계')&(dfs['동']!='소계')&(dfs['동']!='합계')&(dfs['기간']!='기간'))]
dfs
```

```python
#데이터를 열어보니 '-' 값이 있어 string으로 인식이 되어 dropna가 되지 않아 '-' value를 int 0으로 변환

dfs = dfs.replace(to_replace={"-":0})
```

```python
dfs['기간'].unique()
```

```python
#null value check
dfs.isnull().sum()
```

```python
dfs.info()
```

```python
#datatype 변환 및 int로 변환하기 위해 value에 포함된 ',' 제거 후 int로 변환

dfs['내국인 인구'] = dfs.copy()['내국인 인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['내국인 인구'] = dfs.copy()['내국인 인구'].astype(float)
dfs['외국인 인구'] = dfs.copy()['외국인 인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['외국인 인구'] = dfs.copy()['외국인 인구'].astype(float)
dfs['세대당인구'] = dfs.copy()['세대당인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['세대당인구'] = dfs.copy()['세대당인구'].astype(float)
dfs['65세이상 내국인 인구'] = dfs.copy()['65세이상 내국인 인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['65세이상 내국인 인구'] = dfs.copy()['65세이상 내국인 인구'].astype(float)
dfs['65세이상 외국인 인구'] = dfs.copy()['65세이상 외국인 인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['65세이상 외국인 인구'] = dfs.copy()['65세이상 외국인 인구'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['65세이상 외국인 인구'] = dfs.copy()['65세이상 외국인 인구'].astype(float)
dfs['총 수급자'] = dfs.copy()['총 수급자'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['총 수급자'] = dfs.copy()['총 수급자'].astype(float)
dfs['인구밀도'] = dfs.copy()['인구밀도'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['인구밀도'] = dfs.copy()['인구밀도'].astype(float)
dfs['독거노인 합계'] = dfs.copy()['독거노인 합계'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['독거노인 합계'] = dfs.copy()['독거노인 합계'].astype(float)
dfs['출생'] = dfs.copy()['출생'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['출생'] = dfs.copy()['출생'].astype(float)
dfs['사망'] = dfs.copy()['사망'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dfs['사망'] = dfs.copy()['사망'].astype(float)
```

```python
dfs
```

```python
#dfsg DF를 group by를 통해 기간-자치구-동이 중복될시 mean값으로 변환
dfsg =dfs.groupby(['기간','자치구','동'], as_index=False).mean()
```
