# 상승률이 높은 4개구의 시각화 

강남구, 서초구, 용산구의 경우 더 높은 가격분포를 갖고 있기때문에 3구와 3구를 제외한 나머지구를 나눠 예측모형을 만든다

```python
#4구 동별 가격 편차 확인

plt.figure(figsize=(8,8))
sns.boxplot(x="자치구", y="거래금액(만원)", data=df4gu)
plt.ylabel('거래금액', fontsize=12)
plt.xlabel('4개구 구', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("4구별 거래금액 편차", fontsize=15)
plt.show()
```

```python
#변수 상관관계 확인

corr_order=df4gu.corr().loc[:,'거래금액(만원)'].abs().sort_values(ascending=False)
corr_order
```

```python
#0.1이상 변수만 따로 DF로 저장

corrcoef = corr_order[corr_order>0.2]
corrcoef = pd.DataFrame(corrcoef)
corrcoef.index.name= 'labels'
corrcoef['index'] = range(1, len(corrcoef) + 1)
corrcoef['labels']= corrcoef.index
corrcoef = corrcoef.set_index('index')
corrcoef.rename(columns={'거래금액(만원)':'values'}, inplace = True)
corrcoef
```

```python
#히트맵으로 상관관계 시각화

cols_to_use = corrcoef.labels.tolist()

temp_df4gu = df4gu[cols_to_use]
corrmat = temp_df4gu.corr(method='spearman')
f, ax = plt.subplots(figsize=(10,10))

sns.heatmap(corrmat, vmax=1., square=True)
plt.title("4구 상관관계 히트맵 상관관계 (corr>0.2)", fontsize=18)
plt.show()
```

```python
#전용면적 by 금액으로 분포도 시각화

col = "전용면적(㎡)"
ulimit = np.percentile(df4gu[col].values, 99.5)
llimit = np.percentile(df4gu[col].values, 0.5)
df4gu[col].loc[df4gu[col]>ulimit] = ulimit
df4gu[col].loc[df4gu[col]<llimit] = llimit

plt.figure(figsize=(18,18))
sns.jointplot(x=df4gu['전용면적(㎡)'].values, y=df4gu['거래금액(만원)'].values,kind='reg',joint_kws={'line_kws':{'color':'red'}}, size=10)
plt.ylabel('거래금액', fontsize=12)
plt.xlabel('전용면적', fontsize=12)
plt.title("4구 전용면적 by 거래금액 (회귀선)", fontsize=15)
plt.show()
```

```python
#거래 금액 연도별 변동 및 표준편차
import math
df4gu_mean = df4gu.groupby(by = '연도').agg({'거래금액(만원)' : 'mean'})
df4gu_mean = list(df4gu_mean['거래금액(만원)'])

df4gu_year = df4gu['연도'].unique()

df4gu_len = len(df4gu_year)

df4gu_std = []

for i in range(df4gu_len):
    year = df4gu_year[i]
    tmp = list(df4gu[df4gu['연도'] == year]['거래금액(만원)'])
    tmp_mean = sum(tmp)/len(tmp)
    tmp2 = []
    for j in tmp:
        tmp2.append((j - tmp_mean)**2)
    df4gu_std.append(math.sqrt(sum(tmp2) / len(tmp)))

df4gu_reliability = []

for i in df4gu_std:
    df4gu_reliability.append(1.96 * i / math.sqrt(len(tmp)))

upper_ci = []
lower_ci = []

for i in range(len(df4gu_mean)):
    upper_ci.append(df4gu_mean[i] + df4gu_reliability[i])
    lower_ci.append(df4gu_mean[i] - df4gu_reliability[i])

    
fig, ax = plt.subplots()
fig.set_size_inches(7, 7)


ax.fill_between(df4gu_year, upper_ci, lower_ci, facecolor='green', interpolate=True, alpha = 0.2)

ax.plot(df4gu_year, df4gu_mean, 'g')

ax.set_xlabel('연도', fontsize = 12)
ax.set_ylabel('거래금액(만원)', fontsize = 12)

ax.tick_params(axis = 'x', labelsize = 12)
ax.tick_params(axis = 'y', labelsize = 12)
plt.title("연도별 평균 거래금액 및 표준편차", fontsize=12)

ax.set_xticks(df4gu_year);
```

# 상관관계가 높은 변수 추출 
# Foward Selection, Backward Elimination, Stepwise Selection

```python
import statsmodels.api as sm
X4gu = df4gu1.drop(['거래금액(만원)','시','주소'], axis=1)
y4gu = df4gu1[('거래금액(만원)')]

Xrest = dfrest1.drop(['거래금액(만원)','시','주소'], axis=1)
yrest = dfrest1[('거래금액(만원)')]
```

```python
def forward_selection(data, target,significance_level=0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features

sw4gufs = forward_selection(X4gu,y4gu)
swrestfs = forward_selection(Xrest,yrest)
print(sw4gu)
print(swrest)
```

```python
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break 
    return features

sw4gube = backward_elimination(X4gu,y4gu)
swrestbe = backward_elimination(Xrest,yrest)
print(sw4gube)
print(swrestbe)
```

```python
#Step-wise Selection

def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break 
        else:
            break
    return best_features
sw4gusw = stepwise_selection(X4gu,y4gu)
swrestsw = stepwise_selection(Xrest,yrest)
print(sw4gusw)
print(swrestsw)
```
